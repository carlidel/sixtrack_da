```python
# Base libraries
import math
import numpy as np
import scipy.integrate as integrate
from tqdm import tqdm
from scipy.special import erf
import pickle
import itertools

# Personal libraries
import sixtrackwrap as sx

import time
import matplotlib.pyplot as plt
```

```python
sample_sizes = [1,10,100,1000]
batch_sizes = [100,1000,10000,100000,1000000]

times = np.zeros((len(sample_sizes), len(batch_sizes)))
```

```python
for i, sample in tqdm(list(enumerate(sample_sizes))):
    for j, batch in tqdm(list(enumerate(batch_sizes))):
        alpha = np.linspace(0, 1, sample)
        coso = sx.radial_scanner(alpha, alpha, alpha, 1.0, 1)
        start = time.time()
        _ = coso.scan(100, 10, batch)
        times[i, j] = time.time() - start
```

```python
times_post = times.copy()
```

```python
times_post
```

```python
for i in range(len(times_post)):
    times_post[i] /= np.sum(times_post[i]) / 10
```

```python
plt.imshow(times_post)
plt.colorbar()
```

```python
sample_sizes = [10000, 100000]
batch_sizes = [100,1000,10000,100000,1000000]

new_times = [[],[]]
for i, sample in tqdm(list(enumerate(sample_sizes))):
    for j, batch in tqdm(list(enumerate(batch_sizes))):
        alpha = np.linspace(0, 1, sample)
        coso = sx.radial_scanner(alpha, alpha, alpha, 1.0, 1)
        start = time.time()
        _ = coso.scan(100, 10, batch)
        new_times[i].append(time.time() - start)
```

```python
new_times = new_times[0][:]
```

```python
new_times
```

```python
times_post
```

```python
new_post = np.append(times_post, [new_times], axis = 0)
```

```python
for i in range(len(new_post)):
    new_post[i] /= np.sum(new_post[i]) / 10
```

```python
plt.imshow(new_post)
plt.colorbar()
```
